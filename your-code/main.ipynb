# Interactive Visualization Lab

Complete the following set of exercises to solidify your knowledge of interactive visualization using Plotly, Cufflinks, and IPyWidgets.

```python
import pandas as pd
import plotly.plotly as py
import cufflinks as cf
from ipywidgets import interact

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'

cf.go_offline()
```

```python
data = pd.read_excel('../data/Online Retail.xlsx')
```

## 1. Create an interactive bar chart showing total quantity and revenue by country (excluding United Kingdom) for the month of April 2011.

```python
data.head()
data.tail()
```

```python
d1 = data[(data.InvoiceDate > '2011-03') & (data.InvoiceDate < '2011-05') & (data.Country != 'United Kingdom') ]
d2 = d1.groupby(['Country'])['Quantity','Revenue'].sum()
d2 = d2.reset_index()
d2
```

```python
d2.iplot(kind='bar',x='Country',xTitle='Countries', yTitle='Total')
```

## 2. Create an interactive line chart showing quantity and revenue sold to France between January 1st and May 31st 2011.

```python
d1 = data[(data.InvoiceDate >= '2011-01') & (data.InvoiceDate < '2011-06') & (data.Country == 'France')]
d1['InvDate'] = pd.to_datetime(d1['InvoiceDate'])
d1['InvDate'] = d1['InvDate'].apply(lambda x:x.date().strftime('20%y-%m-%d'))
d2 = d1.groupby(['InvDate'])['Quantity','Revenue'].sum()
#d2 = d2.reset_index()
d2.head()
```

```python
d2.iplot(kind='line',xTitle='Jan - May', yTitle='France',title='Quantity and Revenue France Jan to May')
```

## 3. Create an interactive scatter plot showing the relationship between average quantity (x-axis) and average unit price (y-axis) for the product PARTY BUNTING with the plot points color-coded by country (categories).

```python
d1 = data[data.Description=='PARTY BUNTING'][['Quantity','UnitPrice','Country']]
d1 = d1.groupby('Country').mean()
d1 = d1.reset_index()
d1.head()
d1.tail()
```

```python
d1.iplot(kind='scatter',x='Quantity',y='UnitPrice', categories='Country',xTitle='Quantity', yTitle='UnitPrice',title='Quantity vs UnitPrice')
```

## 4. Create a set of interactive histograms showing the distributions of quantity per invoice for the following countries: EIRE, Germany, France, and Netherlands.

```python
data.head(2)
countries = ['EIRE', 'Germany', 'France', 'Netherlands']
d1 = data[data.Country.isin(countries)][['Quantity','InvoiceNo','Country']]

d1 = d1.pivot_table(values='Quantity',columns='Country',index='InvoiceNo',aggfunc='sum')
d1.head()
```

```python
d1.iplot(kind='hist' , subplots=True
         ,xTitle='Invoice', yTitle='Quantity',title='Quantity per Invoice')
```

## 6. Create an interactive line chart showing quantity sold by day for the United Kingdom. Add drop-down boxes for Year and Month that allow you to filter the date range that appears in the chart.

```python
data['Year'] = pd.DatetimeIndex(data['InvoiceDate']).year
data['Month'] = pd.DatetimeIndex(data['InvoiceDate']).month
data['Day'] = pd.DatetimeIndex(data['InvoiceDate']).day
uk = data[data['Country']=='United Kingdom']
uk.head()

```

```python
@interact(Year=list(uk['Year'].unique()), 
          Month=list(uk['Month'].unique())
         )


def linechart(Year, Month):
    uk2 = uk[(uk.Year == Year)&(uk.Month == Month)]
    dat = uk2.pivot_table(values='Day', columns=['Year','Month'],
                            index='Quantity', aggfunc='sum').reset_index()
    dat.fillna(0,inplace=True)

    dat.iplot(kind='line', x='Quantity', xTitle='Number', 
               yTitle='Quantity', title='quantity sold by day ')
```

## 7. Create an interactive scatter plot that plots number of invoices (x-axis) vs. number of customers (y-axis) and the plot points represent individual products. Add two sliders that control the x and y axis ranges.

```python
agg_func = {'InvoiceNo':'nunique',
            'Quantity':'sum',
            'UnitPrice':'mean',
            'Revenue':'sum',
            'CustomerID':'nunique'}

products = uk.groupby('Description').agg(agg_func)
```

```python
products.head()
max(products['CustomerID'])

@interact(Customers = (min(products['CustomerID']), max(products['CustomerID'])),
          Invoices = (min(products['InvoiceNo']), max(products['InvoiceNo']))
         )

def scatter(Customers,Invoices):
    data = products[(products['CustomerID']>=Customers) & 
              (products['InvoiceNo']>=Invoices) 
              ]

    data.iplot(kind='scatter',mode='markers', x='InvoiceNo', y='CustomerID', 
               xTitle='Invoices', yTitle='Customers',
               title='Invoices vs. Customers')
```
